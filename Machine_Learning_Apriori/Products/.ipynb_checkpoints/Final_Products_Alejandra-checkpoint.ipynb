### Basket

```python
basket = (finaldf.groupby(['order_id', 'product_name'])['reordered']
          .sum().unstack().reset_index().fillna(0)
          .set_index('order_id'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
basket_sets = basket.applymap(encode_units)
basket_sets.head()
```

### Rules

```python
frequent_itemsets = apriori(basket_sets, min_support=0.003, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules.dtypes
```

```python
rules_by_conf = rules.sort_values(by="confidence", ascending=True)
max_conf_rule = rules_by_conf['confidence'].max()
min_conf_rule = rules_by_conf['confidence'].min()
avg_conf_rule = rules_by_conf['confidence'].mean()
print(max_conf_rule, min_conf_rule, avg_conf_rule)
```

```python
rules_by_lift = rules.sort_values(by="lift", ascending=False)
max_lift_rule = rules_by_lift['lift'].max()
min_lift_rule = rules_by_lift['lift'].min()
avg_lift_rule = rules_by_lift['lift'].mean()
print(max_lift_rule, min_lift_rule, avg_lift_rule)
```

```python
rules_by_supp = rules.sort_values(by="support", ascending=False)
max_supp_rule = rules_by_supp['support'].max()
min_supp_rule = rules_by_supp['support'].min()
avg_supp_rule = rules_by_supp['support'].mean()
print(max_supp_rule, min_supp_rule, avg_supp_rule)
```

```python
rules_2 = rules[(rules["lift"]>=2.5)& 
     (rules["confidence"]>=0.20)]
# rules_2
```

```python
rules_2_by_lift = rules_2.sort_values(by="lift", ascending=False)
# rules_2
```

```python
def draw_graph(rules_3, rules_to_show):
    G1 = nx.DiGraph()
    color_map=[]
    N = 50
    colors = np.random.rand(N)
    strs=['R0', 'R1', 'R2', 'R3', 'R4', 'R5', 'R6', 'R7', 'R8', 'R9', 'R10', 'R11']
    for i in range(rules_to_show):
        G1.add_nodes_from(["R"+str(i)])
        for a in rules_3.iloc[i]['antecedents']:
            G1.add_nodes_from([a])
            G1.add_edge(a, "R"+str(i), color=colors[i] , weight = 5)
        for c in rules_2.iloc[i]['consequents']:
            G1.add_nodes_from([c])
            G1.add_edge("R"+str(i), c, color=colors[i],  weight=2)
    for node in G1:
        found_a_string = False
        for item in strs:
            if node==item:
                found_a_string = True
        if found_a_string:
            color_map.append('yellow')
        else:
            color_map.append('green')
    edges = G1.edges()
    colors = [G1[u][v]['color'] for u,v in edges]
    weights = [G1[u][v]['weight'] for u,v in edges]
    pos = nx.spring_layout(G1, k=16, scale=1)
    nx.draw(G1, pos, edges=edges, node_color = color_map, edge_color=colors, width=weights, font_size=16, with_labels=False)
    for p in pos:  # raise text positions
        pos[p][1] += 0.07
        nx.draw_networkx_labels(G1, pos)
        plt.title("Product Rules",fontsize=20,fontweight="bold")
        plt.show()
product_graph = draw_graph (rules_3, 8)
# plt.savefig('product_rules.png', bbox_inches="tight")
```
